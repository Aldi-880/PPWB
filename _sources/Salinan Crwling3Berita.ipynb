```python
url = "https://www.antaranews.com/" #menentukan url
```

```python
import requests
from bs4 import BeautifulSoup
import nltk
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser') #Isi teks dari respons HTTP yang diterima dari server web setelah melakukan permintaan GET.
# pages = soup.findAll('article', {'class' :'list-content__item'})
first_page = soup.findAll('li',"dropdown mega-full menu-color1") #menemukan semua list yang berisi link kategori

# print(first_page)
save_categori = []
for links in first_page:
  categori = links.find('a').get('href')
  save_categori.append(categori)
save_categori
```

```python
categori_search = [save_categori[1],save_categori[6],save_categori[7]] #categori yang akan disearch terdapat pada indeks 1,3,6
categori_search
```

```python
datas=[]

for ipages in range(1,17):
  for beritas in categori_search:
    # print(beritas)
    response_berita = requests.get(beritas+"/"+str(ipages))
    namecategori = beritas.split("/")
    soup_berita = BeautifulSoup(response_berita.text, 'html.parser')
    pages_berita = soup_berita.findAll('article', {'class' :'simple-post simple-big clearfix'})

    i = 0
    for items in pages_berita:
      i += 1
      get_link_in = items.find("a").get("href")
      response_artikel = requests.get(get_link_in)
      soup_artikel = BeautifulSoup(response_artikel.text, 'html.parser')

      cek_judul = soup_artikel.findAll("h1","post-title")
      if cek_judul:
        judul = soup_artikel.find("h1","post-title").text
      else:
        judul=""

      label = namecategori[-1]

      try:
        date = soup_artikel.find("span", "article-date").text
      except AttributeError:
          date = "Data tanggal tidak ditemukan"
      cek_baca_juga = soup_artikel.findAll("span","baca-juga")
      trash1 = ""
      if cek_baca_juga:
        for bacas in cek_baca_juga:
          text_trash = bacas.text
          trash1 += text_trash + ' '
      else:
        trash1 = ""

      artikels = soup_artikel.find_all('div', {'class' :'post-content clearfix'})
      if artikels:
        artikel_content = soup_artikel.find('div', {'class' :'post-content clearfix'}).text
        artikel = artikel_content.replace("\n"," ").replace("\t"," ").replace("\r"," ").replace(trash1,"").replace("\xa0","")
      else:
        artikel = ""
      cek_author = soup_artikel.findAll("p","text-muted small mt10")
      if cek_author:
        author = soup_artikel.find("p","text-muted small mt10").text.replace("\t\t"," ")
      else:
        author = ""
      datas.append({
      'Tanggal' : date,
      'Penulis':author,
      'Judul': judul,
      'Artikel': artikel,
      'Label' :  label
          })
```

```python
# Menyimpan data dalam bentuk CSV
import csv
csv_filename = '/content/drive/MyDrive/PPW/DataUAS.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Tanggal', 'Penulis', 'Judul', 'Artikel', 'Label']
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for entry in datas:
        csv_writer.writerow(entry)

print(f"Data telah disimpan dalam file {csv_filename}")
```

```python
import requests
from bs4 import BeautifulSoup
import nltk

# Fungsi untuk melakukan crawling berita dari suatu URL
def crawl_berita(url):
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    judul = soup.find('h1', class_="post-title")
    article = soup.find('div', class_="post-content clearfix")  # Sesuaikan dengan struktur HTML halaman web berita
    if judul and article:
        judul_berita = judul.get_text()
        article_text = article.get_text()
        return judul_berita, article_text
    else:
        return None, None

# Daftar URL dari tiga situs berita yang ingin Anda crawling
urls = [
    "https://www.antaranews.com/berita/3855282/presiden-jokowi-tunjuk-marthinus-hukom-jadi-kepala-bnn",
    "https://www.antaranews.com/berita/3853461/dejan-gloria-paparkan-kunci-juarai-turnamen-super-300-perdana-di-india?utm_source=antaranews&utm_medium=desktop&utm_campaign=menu_news",
    "https://www.antaranews.com/berita/3853809/sasa-ramaikan-event-otomotif-decemblar-kustom-weekender-di-ancol?utm_source=antaranews&utm_medium=desktop&utm_campaign=menu_news"
]

# Loop melalui setiap URL, melakukan crawling, dan tokenisasi kalimat
for url in urls:
    judul_berita, article_text = crawl_berita(url)

    if judul_berita and article_text:
        sentences = nltk.sent_tokenize(article_text)

        # Cetak judul dan kalimat-kalimat dari satu berita
        print(f"Judul Berita dari {url}: {judul_berita}")
        print("Isi Berita:")
        for sentence in sentences:
            print(sentence)
        print("\n")
    else:
        print(f"Gagal mengambil berita dari {url}\n")
```
